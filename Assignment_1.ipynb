# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
plt.figure(figsize=(15,20))
plt.subplot(2,2,1).title.set_text('SL V/S Classification')
sns.boxplot(x="Classification",y='SL',data=df)
plt.subplot(2,2,2).title.set_text('SW V/S Classification')
sns.boxplot(x="Classification",y='SW',data=df)
plt.subplot(2,2,3).title.set_text('PL V/S Classification')
sns.boxplot(x="Classification",y='PL',data=df)
plt.subplot(2,2,4).title.set_text('PW V/S Classification')
sns.boxplot(x="Classification",y='PW',data=df)
plt.show()
```

Box plot shows that 
Iris setosa SL,PL,PW values are very small compair to other two classification but its SW value is higher than other classes
Iris virginica Sl,PL,PW values are hiher than other classification but its Sw value range come between the other two classes
Iris-versicolor All values except SW came between other two classification's value. its SW value lower than other two classification.
its shows that Sw value of all three classification behave differently to others

```python
plt.figure(figsize=(15,10))
plt.subplot(2,1,1).title.set_text('SW V/S SL Based on Classification')
sns.stripplot(x="SL",y="SW",hue="Classification",data=df)
plt.subplot(2,1,2).title.set_text('PW V/S PL Based on Classification')
sns.stripplot(x="PL",y="PW",hue="Classification",data=df)
plt.show()
```

Sl v/s sw plot shows that for setosa values are in specific range. but other two classification some values mixed in certaiin range
Pl/Pw plot clearly show the difference between the different classification's

```python
plt.figure(figsize=(20,15))
p=sns.pairplot(df,hue="Classification")
p.fig.suptitle("Relation b/w SL,SW,PL,PW",y=1.08)
plt.show()
```

The pairplot shows the relationship between column based on classification

```python
plt.figure(figsize=(10,8))
plt.title("Heat Map of Corelation between SL,SW,PL,PW")
sns.heatmap(df.corr(),annot=True,cmap="Blues")
plt.show()
```

```python
sub=df.groupby("Classification")
Is=sub.get_group('Iris-setosa')
Iv=sub.get_group('Iris-versicolor')
Ig=sub.get_group('Iris-virginica')
```

```python
plt.figure(figsize=(10,8))
plt.title("Iris-setosa")
sns.heatmap(Is.corr(),annot=True,cmap="Blues")
plt.show()
```

The heat map shows the corelation between Sl,SW,PL,PW based on Iris-setosa group
SW & SL show strong relation

```python
plt.figure(figsize=(10,8))
plt.title("Iris-versicolor")
sns.heatmap(Iv.corr(),annot=True,cmap="Blues")
plt.show()
```
